```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

# Define the parameter grid to search
param_grid = {
    'tfidf__max_features': [1000, 5000, 10000],
    'tfidf__ngram_range': [(1, 1), (1, 2)],
    'clf__alpha': [0.1, 0.5, 1.0]
}

# Create tokenization and modelling pipeline
model_0 = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", MultinomialNB())
])

# Create GridSearchCV instance
grid_search = GridSearchCV(estimator=model_0, param_grid=param_grid, cv=5, scoring='accuracy', verbose=1)

# Fit GridSearchCV to the training data
grid_search.fit(train_sentences, train_labels)

# Get the best score and parameters
best_score = grid_search.best_score_
best_params = grid_search.best_params_

print("Best Score:", best_score)
print("Best Parameters:", best_params)
```

```python
# Function to evaluate: accuracy, precision, recall, f1-score
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def calculate_results(y_true, y_pred):
  """
  Calculates model accuracy, precision, recall and f1 score of a binary classification model.

  Args:
  -----
  y_true = true labels in the form of a 1D array
  y_pred = predicted labels in the form of a 1D array

  Returns a dictionary of accuracy, precision, recall, f1-score.
  """
  # Calculate model accuracy
  model_accuracy = accuracy_score(y_true, y_pred) * 100
  # Calculate model precision, recall and f1 score using "weighted" average
  model_precision, model_recall, model_f1, _ = precision_recall_fscore_support(y_true, y_pred, average="weighted")
  model_results = {"accuracy": model_accuracy,
                  "precision": model_precision,
                  "recall": model_recall,
                  "f1": model_f1}
  return model_results
```

```python
# We can use this encoding layer in place of our text_vectorizer and embedding layer
sentence_encoder_layer = hub.KerasLayer("https://tfhub.dev/google/universal-sentence-encoder/4",
                                        input_shape=[], # shape of inputs coming to our model
                                        dtype=tf.string, # data type of inputs coming to the USE layer
                                        trainable=False, # keep the pretrained weights (we'll create a feature extractor)
                                        name="USE")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
# Create model using the Sequential API
model_3 = tf.keras.Sequential([
  sentence_encoder_layer, # take in sentences and then encode them into an embedding
  layers.Dense(64, activation="relu"),
  layers.Dense(1, activation="sigmoid")
], name="model_6_USE")

# Compile model
model_3.compile(loss="binary_crossentropy",
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])

early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

model_3.summary()
```

```python
# Train a classifier on top of pretrained embeddings
model_3_history = model_3.fit(train_sentences,
                              train_labels,
                              epochs=50,
                              validation_data=(val_sentences, val_labels),
                              callbacks=[early_stopping,create_tensorboard_callback(SAVE_DIR,
                                                                     "tf_hub_sentence_encoder")])
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.callbacks import EarlyStopping

# Define input layer for sentences
sentence_input = layers.Input(shape=[], dtype=tf.string, name="sentence_input")

# Load the sentence encoder layer (assuming 'sentence_encoder_layer' is defined)
sentence_encoder_layer = sentence_encoder_layer

# Encode the input sentences
sentence_embedding = sentence_encoder_layer(sentence_input)

# Add additional layers for complexity
x = layers.Dense(512, activation="relu")(sentence_embedding)  # Increased units
x = layers.Dense(256, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)  # L2 regularization
x = layers.Dense(128, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)  # L2 regularization

# Add output layer for binary classification
output = layers.Dense(1, activation="sigmoid", name="output")(x)

# Construct the model using Functional API
model_4 = Model(inputs=sentence_input, outputs=output, name="model_4_USE")

# Compile the model with different learning rate
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)  # Adjusted learning rate
model_4.compile(loss="binary_crossentropy",
                optimizer=optimizer,
                metrics=["accuracy"])

# Define early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)  # Increased patience

# Print model summary
model_4.summary()

# Train the model with more epochs
model_4_history = model_4.fit(train_sentences,
                              train_labels,
                              epochs=150,  # Increased epochs
                              validation_data=(val_sentences, val_labels),
                              callbacks=[early_stopping, create_tensorboard_callback(SAVE_DIR, "tf_hub_sentence_encoder")])
```
